```python
import numpy as np
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
df =pd.read_csv("/content/Churn_Modelling.csv")
df.head()
```

```python
# split the data into dependent and independent varaible
```

```python
x= df.iloc[:,3:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
x=pd.get_dummies(x,columns=['Geography','Gender'])
x
```

```python
x=x.drop(['Geography_France','Gender_Female'],axis=1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_stand =sc.fit_transform(x)
x_stand
```

```python
# split the data into training and testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_stand,y,test_size=0.2,random_state=0)
```

## Building Deep Neural Network(MultilayerPerceptron)

```python
from tensorflow import keras
```

```python
# Create API
dnn =tf.keras.models.Sequential()
# Adding hidden layer
dnn.add(tf.keras.layers.Dense(units=6,activation='relu'))
# Adding Second hidden layer
dnn.add(tf.keras.layers.Dense(units=6,activation='relu'))

# Adding output
dnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

# Train the model
dnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"])

# apply training dataset and fit the model
dnn.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
# Predict the test result
y_pred =dnn.predict(x_test)
```

```python
y_pred =(y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
