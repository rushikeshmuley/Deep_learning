```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.utils import img_to_array,array_to_img,load_img
```

```python
img = load_img("/content/cat.jpg")
```

```python
img
```

```python
x = img_to_array(img)
```

```python
x
```

```python
x =x.reshape((1, ) + x.shape)
```

```python
x
```

# Data Augmentation

```python
augmentation = ImageDataGenerator(rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.3,
    shear_range=0.55,     # Shear means Streach
    zoom_range=0.22,
    horizontal_flip=True,
    vertical_flip=True,
    rescale=1./255,)
```

```python
augmentation
```

```python
dir_1 = '/content/sample_data'
i =0
for batch in augmentation.flow(x,batch_size=1,save_prefix ='image',save_to_dir=dir_1,save_format='jpeg'):
  i +=1
  if i >20:
    break
```
